```python
class Convolution(object):
    def __init__(self, src):
        self.f_para = np.fft.fft(src)

    def __mul__(self, x):
        y = self.f_para * np.fft.fft(x)
        y = np.fft.ifft(y)
        return y.real
    def _apply_adj(self, x):
        y = self.f_para.conjugate() * np.fft.fft(x)
        y = np.fft.ifft(y)
        return y.real
```

```python
nsmp = 1024
dt = 0.001
T = (nsmp-1)*dt
xt = np.linspace(0, T, nsmp)
a = 20.0
fo = signal.ricker(nsmp, a)
fo1 = copy.deepcopy(fo)
# fo1[np.where(fo1<0)] = 0


f1 = np.roll(fo1, -400)
for i in range(len(f1)):
    if f1[i] < 0:
        f1[i]=0
A = Convolution(f1)
dv = np.zeros(nsmp)
dv[250] = 0.1
dv[500] = 0.5
dv[800] = -0.5
fig,ax = plt.subplots(1,3,figsize=[12,3])
ax[0].plot(f1)
ax[1].plot(dv)
d = A._apply_adj(A*dv)
dobs = A*dv
ax[2].plot(dobs)
```

```python
def misfit(dv,do,A,OT_info):
    dp = A * dv
    t_mode = OT_info['t_mode']
    c = OT_info['c']
    a = OT_info['a']
    p = OT_info['p']
    res, adj, obj= optimal_transport_fwi(do, dp, 1, transform_mode=t_mode, c_ratio=c, exp_a=a, env_p=p, npad=0)
    
    gradient = A._apply_adj(adj)
    
    return 0.5*obj, 0.5*gradient
    
```

```python
dv0 = np.zeros(nsmp)
OT_info=dict()
# OT_info['t_mode'] = 'quadratic'
# OT_info['c'] = 1
OT_info['t_mode'] = 'linear'
OT_info['c'] = 0
OT_info['a'] = 1
OT_info['p'] = 2

dv0[250] = 0.1

obj,grad = misfit(dv0,dobs,A,OT_info)

obj2, grad2 = misfit(dv0-grad*1e-12,dobs,A,OT_info)
print(obj, obj2)
# plt.plot(grad)
fig,ax = plt.subplots(1,2,figsize=[12,3])
ax[0].plot(grad)
ax[1].plot(dv0)
```

```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self
 
  def __next__(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration
 
myclass = MyNumbers()
myiter = iter(myclass)
 
for x in myiter:
  print(x)
```

```python
import sys
 
def fibonacci(n): # 生成器函数 - 斐波那契
    a, b, counter = 0, 1, 0
    while True:
        if (counter > n): 
            return
        yield a
        a, b = b, a + b
        counter += 1
f = fibonacci(10) # f 是一个迭代器，由生成器返回生成
 
while True:
    try:
        print (next(f), end=" ")
    except StopIteration:
        sys.exit()
```

```python
from functools import wraps
 
def logit(logfile='out.log'):
    def logging_decorator(func):
        @wraps(func)
        def wrapped_function(*args, **kwargs):
            log_string = func.__name__ + " was called"
            print(log_string)
            # 打开logfile，并写入内容
            with open(logfile, 'a') as opened_file:
                # 现在将日志打到指定的logfile
                opened_file.write(log_string + '\n')
            return func(*args, **kwargs)
        return wrapped_function
    return logging_decorator
 
@logit()
def myfunc1():
    pass
 
myfunc1()
# Output: myfunc1 was called
# 现在一个叫做 out.log 的文件出现了，里面的内容就是上面的字符串
 
@logit(logfile='func2.log')
def myfunc2():
    pass
 
myfunc2()
```
